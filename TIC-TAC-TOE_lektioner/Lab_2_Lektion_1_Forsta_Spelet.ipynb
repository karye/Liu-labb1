## 📖 Del 1 – Vad är Tre i rad?

**Tre i rad** (engelska: *Tic-Tac-Toe*) är ett av världens äldsta spel.
Arkeologer har hittat spelbräden ristade i sten i det antika Egypten – spelet spelades redan **3 500 år sedan**! 🏛️

Spelet är enkelt att lära sig men svårt att bemästra. Det spelas av:
- ✖️ **Spelare X** – den som börjar
- ⭕ **Spelare O** – den som svarar

Målet är att få **tre av sina symboler på rad** – vågrätt, lodrätt eller diagonalt.

## 🎯 Del 2 – Spelreglerna

### Spelplanen
Spelplanen är ett **3×3 rutnät** med 9 rutor:

```
+---+---+---+
| 1 | 2 | 3 |
+---+---+---+
| 4 | 5 | 6 |
+---+---+---+
| 7 | 8 | 9 |
+---+---+---+
```

### Så här spelar man
1. Spelare **X** och **O** turas om att placera sin symbol i en tom ruta
2. Den som först får **tre i rad** vinner (vågrätt, lodrätt eller diagonalt)
3. Om alla 9 rutor är fyllda utan vinnare → **oavgjort**

### Exempel på ett avslutat spel
```
| X | O | X |
| O | X | . |
| . | . | O |
```
*Vem vinner här? (Tips: kolla diagonalen!) 🤔*

### En annan ställning
```
| X | X | X |
| O | O | . |
| . | . | . |
```
*X vinner – tre i rad på översta raden!* ✅

## 🔢 Del 3 – Cellerna i koden

En dator kan inte se symbolerna X och O – den jobbar med **siffror**.
Vi gör en enkel översättning:

| Siffra | Betydelse | Symbol |
|--------|-----------|--------|
| `0`    | Tom ruta  | `.`    |
| `1`    | Spelare X | `X`    |
| `2`    | Spelare O | `O`    |

### Spelplanen som en lista i Python
Vi representerar 3×3-rutnätet som en **lista av listor** (kallas 2D-lista eller matris):

```python
# En tom spelplan
board = [
    [0, 0, 0],   # Rad 0 (översta raden)
    [0, 0, 0],   # Rad 1 (mellanraden)
    [0, 0, 0],   # Rad 2 (nedersta raden)
]
```

För att komma åt en specifik ruta skriver vi `board[rad][kolumn]`.
Indexen börjar på **0**, inte 1!

```
        kol 0  kol 1  kol 2
rad 0  [  X  ][  O  ][  .  ]
rad 1  [  O  ][  X  ][  .  ]
rad 2  [  .  ][  .  ][  O  ]
```


````python
# ============================================================
# Exempel: Spelplan representerad som en lista av listor
# ============================================================

# Spelplanens storlek (3x3)
BOARD_SIZE = 3

# Skapa spelplanen:
#   0 = tom ruta
#   1 = Spelare X
#   2 = Spelare O
board_exempel = [
    [1, 2, 1],   # Rad 0: X, O, X
    [2, 1, 0],   # Rad 1: O, X, tom
    [0, 0, 2],   # Rad 2: tom, tom, O
]

# Skriv ut hela planen
print("Spelplanen som en lista av listor:")
print(board_exempel)
print()

# Kom åt enskilda rutor
print("Rutan uppe till vänster  (rad 0, kol 0):", board_exempel[0][0], "→ X")
print("Rutan i mitten           (rad 1, kol 1):", board_exempel[1][1], "→ X")
print("Rutan nere till höger    (rad 2, kol 2):", board_exempel[2][2], "→ O")
print()

# Loopa igenom hela planen för att skriva ut den snyggt
symboler = {0: ".", 1: "X", 2: "O"}  # Ordbok: siffra → symbol
print("Spelplanen som ett rutnät:")
for rad in range(BOARD_SIZE):          # Gå igenom varje rad
    rad_text = " | ".join(symboler[board_exempel[rad][kol]] for kol in range(BOARD_SIZE))
    print(f"  {rad_text}")
````

## 🏆 Del 4 – Vinnare-logiken

Det finns exakt **8 möjliga sätt** att vinna i Tre i rad:

### 3 vågräta rader
```
→ → →      . . .      . . .
. . .      → → →      . . .
. . .      . . .      → → →
```

### 3 lodräta kolumner
```
↓ . .      . ↓ .      . . ↓
↓ . .      . ↓ .      . . ↓
↓ . .      . ↓ .      . . ↓
```

### 2 diagonaler
```
↘ . .      . . ↙
. ↘ .      . ↙ .
. . ↘      ↙ . .
```

**Totalt: 3 + 3 + 2 = 8 vinnande linjer** 🎯

I koden kollar vi dessa 8 linjer en i taget.
Om alla tre cellerna i en linje tillhör **samma spelare** (och inte är tomma) → den spelaren vinner!


````python
# ============================================================
# Funktionen check_winner – kollar om det finns en vinnare
# ============================================================

BOARD_SIZE = 3  # Spelplanens storlek (3x3)

def check_winner(board):
    """
    Kollar om det finns en vinnare på spelplanen.

    Argument:
        board: Spelplanen (2D-lista 3x3 med värden 0, 1 eller 2)

    Returnerar:
        1    → Spelare 1 (X) vinner
        2    → Spelare 2 (O) vinner
        3    → Oavgjort (alla rutor fyllda, ingen vinnare)
        None → Spelet pågår fortfarande
    """

    # ----------------------------------------------------------
    # STEG 1: Kolla alla vågräta rader (3 stycken)
    # ----------------------------------------------------------
    for rad in range(BOARD_SIZE):        # rad = 0, 1, 2
        # Kolla om den första cellen i raden INTE är tom
        if board[rad][0] != 0:
            # Kolla om ALLA celler i raden är lika med den första
            # all(...) returnerar True om villkoret stämmer för alla element
            if all(board[rad][k] == board[rad][0] for k in range(BOARD_SIZE)):
                return board[rad][0]     # Returnera spelaren (1 eller 2)

    # ----------------------------------------------------------
    # STEG 2: Kolla alla lodräta kolumner (3 stycken)
    # ----------------------------------------------------------
    for kol in range(BOARD_SIZE):        # kol = 0, 1, 2
        # Kolla om den första cellen i kolumnen INTE är tom
        if board[0][kol] != 0:
            # Kolla om ALLA celler i kolumnen är lika med den första
            if all(board[k][kol] == board[0][kol] for k in range(BOARD_SIZE)):
                return board[0][kol]     # Returnera spelaren

    # ----------------------------------------------------------
    # STEG 3: Kolla diagonalen uppifrån-vänster → ner-höger (↘)
    #         Cellerna: (0,0), (1,1), (2,2)
    # ----------------------------------------------------------
    if board[0][0] != 0:
        # all(board[k][k] ...) kollar (0,0), (1,1), (2,2)
        if all(board[k][k] == board[0][0] for k in range(BOARD_SIZE)):
            return board[0][0]

    # ----------------------------------------------------------
    # STEG 4: Kolla diagonalen uppifrån-höger → ner-vänster (↙)
    #         Cellerna: (0,2), (1,1), (2,0)
    # ----------------------------------------------------------
    if board[0][BOARD_SIZE - 1] != 0:
        # board[k][BOARD_SIZE-1-k] ger: k=0→kol2, k=1→kol1, k=2→kol0
        if all(board[k][BOARD_SIZE - 1 - k] == board[0][BOARD_SIZE - 1]
               for k in range(BOARD_SIZE)):
            return board[0][BOARD_SIZE - 1]

    # ----------------------------------------------------------
    # STEG 5: Kolla om det är oavgjort
    #         Oavgjort = inga tomma rutor kvar OCH ingen vinnare
    # ----------------------------------------------------------
    if all(board[i][j] != 0 for i in range(BOARD_SIZE) for j in range(BOARD_SIZE)):
        return 3                         # 3 = oavgjort

    # ----------------------------------------------------------
    # STEG 6: Ingen vinnare och tomma rutor finns → spelet pågår
    # ----------------------------------------------------------
    return None


def visa_brade(board):
    """Visar spelplanen i ett snyggt, läsbart format."""
    symboler = {0: ".", 1: "X", 2: "O"}  # Översätt siffror till symboler
    print("Spelplanen:")
    print("  +---+---+---+")
    for rad in range(BOARD_SIZE):
        rad_text = " | ".join(symboler[board[rad][kol]] for kol in range(BOARD_SIZE))
        print(f"  | {rad_text} |")
        print("  +---+---+---+")
    print()

print("✅ Funktionerna check_winner och visa_brade är laddade!")
````

````python
# ============================================================
# Övning 1.5 – Förstå check_winner med steg-för-steg-loggning
# ============================================================
# Vi lägger till utskrifter i en kopia av check_winner
# för att se exakt vad funktionen gör.

def check_winner_debug(board):
    """En version av check_winner som skriver ut vad den gör."""

    # Kolla rader
    for rad in range(BOARD_SIZE):
        if board[rad][0] != 0 and all(board[rad][k] == board[rad][0] for k in range(BOARD_SIZE)):
            print(f"  ✅ Vinnare hittad på RAD {rad}!")
            return board[rad][0]
        else:
            print(f"  ⬜ Rad {rad}: ingen vinnare")

    # Kolla kolumner
    for kol in range(BOARD_SIZE):
        if board[0][kol] != 0 and all(board[k][kol] == board[0][kol] for k in range(BOARD_SIZE)):
            print(f"  ✅ Vinnare hittad på KOLUMN {kol}!")
            return board[0][kol]
        else:
            print(f"  ⬜ Kolumn {kol}: ingen vinnare")

    # Kolla diagonaler
    if board[0][0] != 0 and all(board[k][k] == board[0][0] for k in range(BOARD_SIZE)):
        print("  ✅ Vinnare hittad på DIAGONAL ↘!")
        return board[0][0]
    else:
        print("  ⬜ Diagonal ↘: ingen vinnare")

    if board[0][BOARD_SIZE-1] != 0 and all(board[k][BOARD_SIZE-1-k] == board[0][BOARD_SIZE-1] for k in range(BOARD_SIZE)):
        print("  ✅ Vinnare hittad på DIAGONAL ↙!")
        return board[0][BOARD_SIZE-1]
    else:
        print("  ⬜ Diagonal ↙: ingen vinnare")

    if all(board[i][j] != 0 for i in range(BOARD_SIZE) for j in range(BOARD_SIZE)):
        print("  🤝 Alla rutor fyllda – oavgjort!")
        return 3

    print("  ⏳ Spelet pågår...")
    return None

# Kör debug-versionen
test_board = [
    [1, 2, 1],
    [2, 1, 0],
    [0, 0, 2],
]
visa_brade(test_board)
print("Steg-för-steg-analys:")
resultat = check_winner_debug(test_board)
print(f"\nSlutresultat: {resultat}")
````

````python
# ============================================================
# Övning 1.8 – Utmaning: Spela ett manuellt spel
# ============================================================
# Simulera ett komplett spel steg för steg.
# Varje drag representeras av (rad, kol, spelare).

def gor_drag(board, rad, kol, spelare):
    """Placerar spelarens symbol på spelplanen."""
    if board[rad][kol] != 0:
        print(f"❌ Rutan ({rad},{kol}) är redan fylld!")
        return False
    board[rad][kol] = spelare
    return True

# Starta ett nytt spel
spelbrade = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0],
]

# Lista med drag: (rad, kolumn, spelare)
# 1 = X, 2 = O
drag = [
    (1, 1, 1),   # X spelar i mitten
    (0, 0, 2),   # O spelar uppe till vänster
    (0, 2, 1),   # X spelar uppe till höger
    (2, 0, 2),   # O spelar nere till vänster
    (0, 1, 1),   # X spelar översta mitten
    # TODO: Lägg till fler drag och se vad som händer!
]

# Genomför alla drag
for i, (rad, kol, spelare) in enumerate(drag):
    symbol = "X" if spelare == 1 else "O"
    print(f"--- Drag {i+1}: Spelare {symbol} spelar på ({rad},{kol}) ---")
    gor_drag(spelbrade, rad, kol, spelare)
    visa_brade(spelbrade)

    # Kolla om spelet är slut
    vinnare = check_winner(spelbrade)
    if vinnare == 1:
        print("🏆 Spelare X vinner!")
        break
    elif vinnare == 2:
        print("🏆 Spelare O vinner!")
        break
    elif vinnare == 3:
        print("🤝 Oavgjort!")
        break
````
