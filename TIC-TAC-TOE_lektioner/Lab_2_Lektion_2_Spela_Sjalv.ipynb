````python
# ============================================================
# Importera bibliotek som behövs för den interaktiva spelplanen
# ============================================================

import ipywidgets as widgets           # Knappar, etiketter och layouter
from IPython.display import display, clear_output  # Visa widgets och rensa utskrifter

# Spelplanens storlek (3x3)
BOARD_SIZE = 3

# ----------------------------------------------------------
# Kopierad från Lektion 1: check_winner
# Vi behöver den för att avgöra om någon vinner!
# ----------------------------------------------------------
def check_winner(board):
    """
    Kollar om det finns en vinnare på spelplanen.
    Returnerar: 1 (X vinner), 2 (O vinner), 3 (oavgjort), None (pågår)
    """
    # Kolla rader
    for rad in range(BOARD_SIZE):
        if board[rad][0] != 0 and all(board[rad][k] == board[rad][0] for k in range(BOARD_SIZE)):
            return board[rad][0]
    # Kolla kolumner
    for kol in range(BOARD_SIZE):
        if board[0][kol] != 0 and all(board[k][kol] == board[0][kol] for k in range(BOARD_SIZE)):
            return board[0][kol]
    # Kolla diagonal ↘
    if board[0][0] != 0 and all(board[k][k] == board[0][0] for k in range(BOARD_SIZE)):
        return board[0][0]
    # Kolla diagonal ↙
    if board[0][BOARD_SIZE-1] != 0 and all(board[k][BOARD_SIZE-1-k] == board[0][BOARD_SIZE-1] for k in range(BOARD_SIZE)):
        return board[0][BOARD_SIZE-1]
    # Oavgjort?
    if all(board[i][j] != 0 for i in range(BOARD_SIZE) for j in range(BOARD_SIZE)):
        return 3
    return None

def visa_brade(board):
    """Visar spelplanen i ett snyggt textformat."""
    symboler = {0: ".", 1: "X", 2: "O"}
    print("Spelplanen:")
    print("  +---+---+---+")
    for rad in range(BOARD_SIZE):
        rad_text = " | ".join(symboler[board[rad][kol]] for kol in range(BOARD_SIZE))
        print(f"  | {rad_text} |")
        print("  +---+---+---+")
    print()

print("✅ Alla bibliotek och funktioner är laddade och klara!")
````

## 🕹️ Del 1 – Spellogiken: make_move

I Lektion 1 lärde vi oss *hur spelplanen ser ut* i koden.
Nu ska vi lära oss *hur spelare gör drag*.

### Vad är ett drag?
Ett drag i Tre i rad är:
1. En spelare väljer en **tom ruta** (rad + kolumn)
2. Deras symbol placeras i den rutan
3. Det är nästa spelares tur

I Python skapar vi en funktion `make_move(board, row, col, player)` som:
- Tar spelplanen, rad, kolumn och spelarens nummer som argument
- **Kollar** om rutan är tom
- **Placerar** symbolen om rutan är tom
- **Returnerar** `True` (lyckades) eller `False` (rutan var fylld)

```
make_move(board, 1, 1, 1)
             ↑    ↑  ↑  ↑
          board  rad kol spelare
```


````python
# ============================================================
# Funktionen make_move – gör ett drag på spelplanen
# ============================================================

def make_move(board, row, col, player):
    """
    Gör ett drag på spelplanen.

    Argument:
        board:  Spelplanen (2D-lista 3x3)
        row:    Vilken rad (0, 1 eller 2)
        col:    Vilken kolumn (0, 1 eller 2)
        player: Vem spelar (1=X, 2=O)

    Returnerar:
        True  → Draget lyckades (rutan var tom)
        False → Draget misslyckades (rutan var redan fylld)
    """

    # ----------------------------------------------------------
    # STEG 1: Kontrollera att rutan är tom
    # board[row][col] == 0 betyder att rutan är tom
    # ----------------------------------------------------------
    if board[row][col] != 0:
        # Rutan är INTE tom – draget är ogiltigt
        print(f"  ⚠️  Rutan ({row},{col}) är redan fylld med {board[row][col]}!")
        return False   # Returnera False = misslyckades

    # ----------------------------------------------------------
    # STEG 2: Rutan är tom – placera spelarens symbol
    # Vi sätter värdet i spelplanen till spelarens nummer
    # ----------------------------------------------------------
    board[row][col] = player   # Sätt 1 (X) eller 2 (O) i rutan

    # ----------------------------------------------------------
    # STEG 3: Returnera True = draget lyckades
    # ----------------------------------------------------------
    return True


# ============================================================
# Testa make_move
# ============================================================

# Starta med en tom spelplan
spelplan = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0],
]

print("=== Spelplanen INNAN något drag ===")
visa_brade(spelplan)

# Spelare X (1) spelar i mitten
print("Drag 1: Spelare X spelar på (1, 1)")
lyckades = make_move(spelplan, 1, 1, 1)   # rad=1, kol=1, spelare=1
print(f"  Draget lyckades: {lyckades}")
visa_brade(spelplan)

# Spelare O (2) spelar uppe till vänster
print("Drag 2: Spelare O spelar på (0, 0)")
lyckades = make_move(spelplan, 0, 0, 2)   # rad=0, kol=0, spelare=2
print(f"  Draget lyckades: {lyckades}")
visa_brade(spelplan)

# Försök att spela på en fylld ruta
print("Drag 3: Spelare X försöker spela på (1, 1) igen...")
lyckades = make_move(spelplan, 1, 1, 1)   # Samma ruta som drag 1!
print(f"  Draget lyckades: {lyckades}")
visa_brade(spelplan)
````

## 🔄 Del 2 – Tursystemet

I ett riktigt spel turas spelarna om. Vi håller koll på vems tur det är
med en variabel `current_player`:

```python
current_player = 1   # X börjar alltid
```

**Efter varje lyckat drag:**
1. Kolla om det finns en vinnare med `check_winner()`
2. Om ingen vunnit → byt spelare
3. Byta spelare: `current_player = 2 if current_player == 1 else 1`

### Flödesschema
```
Spelare väljer ruta
       ↓
  Rutan tom?
  /        \
NEJ        JA
 ↓          ↓
Försök   Placera symbol
igen         ↓
         check_winner()
         /    |    \
      Vann  Pågår  Oavgjort
        ↓     ↓      ↓
     Sluta  Byt    Sluta
           spelare
```


````python
# ============================================================
# Experiment – Utforska make_move
# ============================================================
# Kör den här cellen och se vad som händer!

# Starta ett nytt spel
exp_board = [[0]*3 for _ in range(3)]
current_player = 1   # X börjar
spelet_pagår = True

# Fördefinierade drag för experimentet
drag_lista = [
    (0, 0),  # X
    (1, 1),  # O
    (0, 1),  # X
    (2, 0),  # O
    (0, 2),  # X – borde vinna!
]

for drag_nr, (rad, kol) in enumerate(drag_lista):
    if not spelet_pagår:
        print("⛔ Spelet är redan slut!")
        break

    symbol = "X" if current_player == 1 else "O"
    print(f"{'='*40}")
    print(f"Drag {drag_nr+1}: Spelare {symbol} spelar på rad={rad}, kol={kol}")

    # Gör draget
    if make_move(exp_board, rad, kol, current_player):
        visa_brade(exp_board)

        # Kolla om spelet är slut
        vinnare = check_winner(exp_board)
        if vinnare == 1:
            print("🏆 SPELARE X VINNER!")
            spelet_pagår = False
        elif vinnare == 2:
            print("🏆 SPELARE O VINNER!")
            spelet_pagår = False
        elif vinnare == 3:
            print("🤝 OAVGJORT!")
            spelet_pagår = False
        else:
            # Byt spelare
            current_player = 2 if current_player == 1 else 1
            nasta = "X" if current_player == 1 else "O"
            print(f"➡️  Nu är det spelare {nasta}:s tur.")
````

## 🖱️ Del 3 – Interaktiva knappar med ipywidgets

Nu bygger vi ett **riktigt spelgränssnitt** direkt i Jupyter!

| `widgets.VBox` | Vertikal layout | Stapla rader |

### Hur fungerar knappar?
```python
knapp = widgets.Button(description="Klicka mig!")

def när_man_klickar(btn):
    print("Du klickade!")

knapp.on_click(när_man_klickar)   # Koppla funktionen till knappen
display(knapp)                     # Visa knappen
```

Varje knapp på spelplanen representerar **en ruta** i rutnätet.
När man klickar → `on_button_click()` anropas → draget registreras!


````python
# ============================================================
# Det interaktiva spelet – Tre i rad med knappar!
# ============================================================
# Kör den här cellen och spela mot en kompis!

# ----------------------------------------------------------
# Spelstatus – variabler som håller koll på spelet
# ----------------------------------------------------------
current_board = [[0] * BOARD_SIZE for _ in range(BOARD_SIZE)]  # Tom spelplan
current_player = [1]   # Lista så att den kan ändras inuti funktioner (Python-trick)
game_over = [False]    # Håller koll på om spelet är slut

# ----------------------------------------------------------
# Statusetikett – visar vem som spelar och resultatet
# ----------------------------------------------------------
status_label = widgets.Label(
    value="🎮 Spelare X börjar! Klicka på en ruta.",
    style={"font_size": "16px"}
)

# ----------------------------------------------------------
# Skapa 3x3 knappar – en för varje ruta på spelplanen
# ----------------------------------------------------------
def create_board_buttons():
    """Skapar 3x3 knappar och returnerar dem som en 2D-lista."""
    buttons = []
    for rad in range(BOARD_SIZE):
        rad_knappar = []           # En lista med knappar för den här raden
        for kol in range(BOARD_SIZE):
            btn = widgets.Button(
                description="",   # Tom från början
                layout=widgets.Layout(
                    width="80px",
                    height="80px"
                ),
                style=widgets.ButtonStyle(font_size="30px")
            )
            # Spara rad och kolumn direkt på knappen
            # (så att on_button_click vet vilken ruta som klickades)
            btn.row = rad
            btn.col = kol
            btn.on_click(on_button_click)   # Koppla klick-funktionen
            rad_knappar.append(btn)
        buttons.append(rad_knappar)
    return buttons


# ----------------------------------------------------------
# Klick-funktionen – anropas när en ruta klickas
# ----------------------------------------------------------
def on_button_click(btn):
    """Hanterar ett knapptryck från spelaren."""

    # Om spelet redan är slut – gör ingenting
    if game_over[0]:
        return

    row = btn.row          # Vilken rad klickades?
    col = btn.col          # Vilken kolumn klickades?
    player = current_player[0]    # Vems tur är det?

    # Försök göra draget
    if make_move(current_board, row, col, player):
        # Draget lyckades!
        symbol = "X" if player == 1 else "O"

        # Uppdatera knappens utseende
        btn.description = symbol
        btn.style.button_color = "lightblue" if player == 1 else "lightyellow"

        # Kolla om spelet är slut
        winner = check_winner(current_board)

        if winner == 3:
            # Oavgjort
            status_label.value = "🤝 Oavgjort! Ingen vann den här gången."
            game_over[0] = True
        elif winner is not None:
            # Någon vann!
            status_label.value = f"🏆 Grattis! Spelare {symbol} vinner! 🎉"
            game_over[0] = True
        else:
            # Spelet pågår – byt spelare
            current_player[0] = 2 if player == 1 else 1
            next_symbol = "X" if current_player[0] == 1 else "O"
            status_label.value = f"🎮 Spelare {next_symbol}:s tur! Klicka på en ruta."
    else:
        # Rutan var fylld – meddela spelaren
        status_label.value = "⚠️  Den rutan är redan fylld! Välj en annan."


# ----------------------------------------------------------
# Återstartsknapp – nollställ allt
# ----------------------------------------------------------
def reset_game(btn):
    """Nollställer spelplanen och startar om."""
    # Nollställ spelplansmatrisen
    for i in range(BOARD_SIZE):
        for j in range(BOARD_SIZE):
            current_board[i][j] = 0

    # Återställ spelvariablerna
    current_player[0] = 1
    game_over[0] = False

    # Återställ statusetiketten
    status_label.value = "🎮 Nytt spel! Spelare X börjar. Klicka på en ruta."

    # Återställ alla knappar
    for rad in board_buttons:
        for b in rad:
            b.description = ""
            b.style.button_color = "white"


# ----------------------------------------------------------
# Bygg gränssnittet och visa det
# ----------------------------------------------------------
board_buttons = create_board_buttons()

# Återstartsknapp
reset_btn = widgets.Button(
    description="🔄 Nytt spel",
    button_style="warning",        # Gul knapp
    layout=widgets.Layout(width="150px", height="40px")
)
reset_btn.on_click(reset_game)

# Visa allt: statusetikett, spelplan (3 rader), återstartsknapp
display(status_label)
display(widgets.VBox([widgets.HBox(rad) for rad in board_buttons]))
display(reset_btn)
````

````python
# ============================================================
# Övning 2.8 – Utmaning: Lägg till en poängvisare i det interaktiva spelet
# ============================================================
# Den här övningen utmanar dig att modifiera det interaktiva spelet
# så att det håller koll på poäng:
#   - X:s vinster
#   - O:s vinster
#   - Antal oavgjorda
#
# Steg:
#   1. Lägg till tre Label-widgets för poängen
#   2. Uppdatera dem i on_button_click när spelet är slut
#   3. Nollställ INTE poängen när reset_game körs (bara spelplanen)

# Poängvariabler
poang = {"X": 0, "O": 0, "Oavgjort": 0}

# Widgets för poängvisning
poang_label_x = widgets.Label(value="X: 0 vinster")
poang_label_o = widgets.Label(value="O: 0 vinster")
poang_label_draw = widgets.Label(value="Oavgjort: 0")
poang_box = widgets.HBox([poang_label_x, poang_label_o, poang_label_draw])

# Ny spelplan och status för det här utökade spelet
poang_board = [[0] * BOARD_SIZE for _ in range(BOARD_SIZE)]
poang_player = [1]
poang_over = [False]
poang_status = widgets.Label(value="🎮 Spelare X börjar! Klicka på en ruta.")

def poang_button_click(btn):
    if poang_over[0]:
        return
    player = poang_player[0]
    if make_move(poang_board, btn.row, btn.col, player):
        symbol = "X" if player == 1 else "O"
        btn.description = symbol
        btn.style.button_color = "lightblue" if player == 1 else "lightyellow"
        winner = check_winner(poang_board)
        if winner == 3:
            poang_status.value = "🤝 Oavgjort!"
            poang_over[0] = True
            # TODO: Öka poang["Oavgjort"] och uppdatera poang_label_draw
        elif winner:
            poang_status.value = f"🏆 Spelare {symbol} vinner!"
            poang_over[0] = True
            # TODO: Öka poang[symbol] och uppdatera rätt poängetikett
        else:
            poang_player[0] = 2 if player == 1 else 1
            ns = "X" if poang_player[0] == 1 else "O"
            poang_status.value = f"🎮 Spelare {ns}:s tur!"

def poang_reset(btn):
    for i in range(BOARD_SIZE):
        for j in range(BOARD_SIZE):
            poang_board[i][j] = 0
    poang_player[0] = 1
    poang_over[0] = False
    poang_status.value = "🎮 Nytt spel! Spelare X börjar."
    for rad in poang_btns:
        for b in rad:
            b.description = ""
            b.style.button_color = "white"
    # Notera: poäng nollställs INTE!

# Skapa knapparna
poang_btns = []
for r in range(BOARD_SIZE):
    rad_btns = []
    for c in range(BOARD_SIZE):
        b = widgets.Button(
            description="",
            layout=widgets.Layout(width="80px", height="80px"),
            style=widgets.ButtonStyle(font_size="30px")
        )
        b.row = r
        b.col = c
        b.on_click(poang_button_click)
        rad_btns.append(b)
    poang_btns.append(rad_btns)

poang_reset_btn = widgets.Button(description="🔄 Nytt spel", button_style="warning",
                                   layout=widgets.Layout(width="150px", height="40px"))
poang_reset_btn.on_click(poang_reset)

display(poang_status)
display(poang_box)
display(widgets.VBox([widgets.HBox(rad) for rad in poang_btns]))
display(poang_reset_btn)
````
