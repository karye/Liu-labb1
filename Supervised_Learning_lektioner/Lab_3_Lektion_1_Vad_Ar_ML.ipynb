## рҹ“– Del 1 вҖ“ Vad ГӨr Machine Learning?

FГ¶restГӨll dig att din e-post automatiskt sorterar bort **spam**. Hur vet datorn att ett mail ГӨr spam?

Den har **lГӨrt sig** frГҘn tusentals exempel!

```
Spam-exempel:         Inte spam:
"Vinn 1 miljon nu!"  "Hej, kan vi ses imorgon?"
"Klicka hГӨr GRATIS"  "MГ¶tet ГӨr kl 14:00"
"SPECIELLT ERBJUDANDE" "GlГ¶m inte ta med lunch"
```

Datorn analyserar **mГ¶nster** i texten och lГӨr sig skilja spam frГҘn vanliga mail.

**Det ГӨr precis vad Machine Learning (ML) ГӨr!**

> рҹӨ– *Machine Learning = Datorer lГӨr sig frГҘn EXEMPEL, istГӨllet fГ¶r att vi programmerar varje regel*

### Tre typer av Machine Learning:

| Typ | FГ¶rklaring | Exempel |
|-----|-----------|---------|
| **Supervised Learning** | Vi ger modellen mГӨrkta exempel (input + rГӨtt svar) | Spam-filter, blomklass |
| **Unsupervised Learning** | Modellen hittar mГ¶nster sjГӨlv, utan mГӨrkta svar | Gruppera kunder |
| **Reinforcement Learning** | Modellen lГӨr sig genom att prova och fГҘ belГ¶ning/straff | Spela schack |

## рҹҺ® Del 5 вҖ“ Interaktiv demo: Se ett dataset!

Nu ska du fГҘ **utforska** hur ett dataset ser ut och hur klassificering fungerar!

KГ¶r cellen nedan och anvГӨnd **slidern** fГ¶r att byta mellan blommor i Iris-datasetet.

````python
# ============================================================
# INTERAKTIV DEMO: Utforska Iris-datasetet
# ============================================================

# Ladda Iris-datasetet
iris = load_iris()
feature_names = ['Sepal lГӨngd (cm)', 'Sepal bredd (cm)', 
                  'Petal lГӨngd (cm)', 'Petal bredd (cm)']
class_names = ['Setosa рҹҢё', 'Versicolor рҹҢә', 'Virginica рҹҢ»']
colors = ['#FF9AA2', '#FFB347', '#87CEEB']

# Skapa en DataFrame fГ¶r enklare hantering
df = pd.DataFrame(iris.data, columns=feature_names)
df['Art'] = [class_names[t] for t in iris.target]

# в”Җв”Җ Interaktiv widget в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җ
output = widgets.Output()

def visa_blomma(blomma_nr):
    with output:
        clear_output(wait=True)
        row = df.iloc[blomma_nr]
        art_index = iris.target[blomma_nr]
        
        print(f"рҹҢё Blomma nummer {blomma_nr + 1} av 150")
        print("=" * 40)
        print(f"  Art: {class_names[art_index]}")
        print()
        print("  MГӨtningar (Features):")
        for fn, val in zip(feature_names, iris.data[blomma_nr]):
            bar = 'в–Ҳ' * int(val * 3)
            print(f"  {fn:22s}: {val:.1f} {bar}")
        print()
        
        # Rita ett stapeldiagram fГ¶r denna blomma
        fig, ax = plt.subplots(figsize=(7, 3))
        bars = ax.bar(feature_names, iris.data[blomma_nr], 
                      color=colors[art_index], edgecolor='black', linewidth=0.5)
        ax.set_title(f'Blomma {blomma_nr + 1}: {class_names[art_index]}', 
                     fontsize=13, fontweight='bold')
        ax.set_ylabel('VГӨrde (cm)')
        ax.set_ylim(0, 8)
        for bar, val in zip(bars, iris.data[blomma_nr]):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
                    f'{val:.1f}', ha='center', va='bottom', fontsize=10)
        plt.xticks(rotation=20, ha='right')
        plt.tight_layout()
        plt.show()

blomma_slider = widgets.IntSlider(
    value=0, min=0, max=149, step=1,
    description='Blomma nr:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='500px')
)

widgets.interactive(visa_blomma, blomma_nr=blomma_slider)
````

---
### рҹ”Қ Interaktiv Scatterplot вҖ“ Se hur arterna skiljer sig!

KГ¶r cellen nedan och **vГӨlj vilka features** du vill jГӨmfГ¶ra med dropdown-menyer.  
Kan du hitta features som tydligt separerar de tre arterna?

````python
# ============================================================
# INTERAKTIV SCATTERPLOT
# ============================================================

output2 = widgets.Output()

def rita_scatter(x_feature, y_feature):
    with output2:
        clear_output(wait=True)
        
        xi = feature_names.index(x_feature)
        yi = feature_names.index(y_feature)
        
        fig, ax = plt.subplots(figsize=(8, 5))
        
        for i, (art, color) in enumerate(zip(class_names, colors)):
            mask = iris.target == i
            ax.scatter(iris.data[mask, xi], iris.data[mask, yi],
                       c=color, label=art, s=60, edgecolors='black', linewidth=0.5)
        
        ax.set_xlabel(x_feature, fontsize=11)
        ax.set_ylabel(y_feature, fontsize=11)
        ax.set_title(f'Iris-datasetet: {x_feature} vs {y_feature}', 
                     fontsize=13, fontweight='bold')
        ax.legend(title='Art', fontsize=10)
        ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        # Tips
        if xi == yi:
            print("вҡ пёҸ VГӨlj OLIKA features pГҘ x- och y-axeln!")
        else:
            setosa_x = iris.data[iris.target==0, xi]
            others_x = iris.data[iris.target!=0, xi]
            overlap = min(setosa_x.max(), others_x.max()) - max(setosa_x.min(), others_x.min())
            if overlap < 0.5:
                print(f"вң… Bra val! {x_feature} skiljer Setosa tydligt frГҘn de andra!")
            else:
                print(f"рҹ’Ў Prova Petal lГӨngd eller Petal bredd fГ¶r bГӨttre separation!")

x_dropdown = widgets.Dropdown(
    options=feature_names, value=feature_names[2],
    description='X-axel:', style={'description_width': 'initial'}
)
y_dropdown = widgets.Dropdown(
    options=feature_names, value=feature_names[3],
    description='Y-axel:', style={'description_width': 'initial'}
)

ui = widgets.VBox([
    widgets.HTML('<b>рҹҺ® VГӨlj features att jГӨmfГ¶ra:</b>'),
    widgets.HBox([x_dropdown, y_dropdown]),
    output2
])

def on_change(change):
    rita_scatter(x_dropdown.value, y_dropdown.value)

x_dropdown.observe(on_change, names='value')
y_dropdown.observe(on_change, names='value')

rita_scatter(feature_names[2], feature_names[3])  # Visa direkt
display(ui)
````
