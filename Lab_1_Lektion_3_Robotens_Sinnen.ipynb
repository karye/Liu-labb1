## ⚙️ Setup – Ladda in världen
Kör den här cellen **först** – den definierar världen och hjälpfunktioner som vi behöver.

````python
# ============================================================
# SETUP – Kör den här cellen först!
# ============================================================

# Celltyper – vad kan finnas i rutnätet?
EMPTY = 0   # Tom cell
WALL  = 1   # Vägg
DIRT  = 2   # Smuts
HOME  = 3   # Robotens hemposition

# Symboltabell – för att skriva ut världen fint
SYMBOLS = {
    EMPTY: ' . ',
    WALL:  ' # ',
    DIRT:  ' ░ ',
    HOME:  ' H ',
}

# En exempelvärld (6x6 rutnät)
# '#' = vägg, '.' = tom, '░' = smuts, 'H' = hem
world = [
    [WALL,  WALL,  WALL,  WALL,  WALL,  WALL],
    [WALL,  HOME,  EMPTY, DIRT,  EMPTY, WALL],
    [WALL,  EMPTY, WALL,  EMPTY, DIRT,  WALL],
    [WALL,  EMPTY, EMPTY, EMPTY, EMPTY, WALL],
    [WALL,  DIRT,  EMPTY, WALL,  EMPTY, WALL],
    [WALL,  WALL,  WALL,  WALL,  WALL,  WALL],
]

# Hemposition (robotens startplats)
home_x = 1
home_y = 1

def visa_världen(world, robot_x, robot_y):
    """Skriver ut rutnätet med robotens position."""
    print("\n  Kolumn: 0    1    2    3    4    5")
    for y, rad in enumerate(world):
        print(f"Rad {y}: ", end="")
        for x, cell in enumerate(rad):
            if x == robot_x and y == robot_y:
                print(' R ', end=" ")  # R = Robot
            else:
                print(SYMBOLS[cell], end=" ")
        print()  # Ny rad
    print()

# Testa att världen ser rätt ut
robot_x = 1  # Robotens startposition
robot_y = 1
print("Välkommen till Lektion 3! Här är vår värld:")
visa_världen(world, robot_x, robot_y)
print("R = Robot, H = Hem, . = Tom, # = Vägg, ░ = Smuts")
````

## Del 1: Vad är perception? 🧠

**Perception** = Vad roboten kan *uppfatta* om sin omgivning just nu.

Tänk på hur du vaknar på morgonen:
- Du *ser* om det är ljust ute ☀️
- Du *känner* om du är hungrig 🍞
- Du *hör* om väckarklockan ringer 📢

En städrobot har liknande frågor:
| Fråga | Svar |
|-------|------|
| Finns det smuts här? | Ja / Nej |
| Är jag hemma? | Ja / Nej |
| Kan jag gå åt höger? | Ja / Nej |

I Python representerar vi **Ja** med `True` och **Nej** med `False`.

## Del 2: Sensor 1 – "Finns det smuts här?" 🧹

Den första sensorn kollar om roboten **står på en cell med smuts**.

Minns: `world[y][x]` ger oss celltypen på rad `y`, kolumn `x`.

````python
# Sensor 1: Kan roboten se smuts på sin position?
# ------------------------------------------------
def can_see_dirt(robot_x, robot_y, world):
    """Returnerar True om det finns smuts på robotens position."""
    cell = world[robot_y][robot_x]  # Hämta celltypen
    return cell == DIRT             # Jämför med DIRT (2)

# Testa sensorn på olika positioner
print("=== Testa can_see_dirt ===")
print(f"Position (1,1) – hemcell:  {can_see_dirt(1, 1, world)}")  # Förväntat: False
print(f"Position (3,1) – smutscell: {can_see_dirt(3, 1, world)}")  # Förväntat: True
print(f"Position (2,1) – tom cell:  {can_see_dirt(2, 1, world)}")  # Förväntat: False

# Visa världen som referens
visa_världen(world, 3, 1)  # Visa roboten på smutscellen
````

## Sensor 2 – "Är jag hemma?" 🏠

Den andra sensorn kollar om roboten **befinner sig på hemposition**.  
Hemposition är där roboten startade: `(home_x, home_y)` = `(1, 1)`.

````python
# Sensor 2: Är roboten hemma?
# ----------------------------
def is_at_home(robot_x, robot_y, home_x, home_y):
    """Returnerar True om roboten är på hemposition."""
    # Kolla om BÅDA koordinaterna stämmer
    return robot_x == home_x and robot_y == home_y

# Testa sensorn
print("=== Testa is_at_home ===")
print(f"Robot på (1,1) – hemma?    {is_at_home(1, 1, home_x, home_y)}")  # Förväntat: True
print(f"Robot på (3,2) – hemma?    {is_at_home(3, 2, home_x, home_y)}")  # Förväntat: False
print(f"Robot på (1,3) – hemma?    {is_at_home(1, 3, home_x, home_y)}")  # Förväntat: False
````

## Sensor 3 – "Kan jag röra mig dit?" 🚶

Den tredje sensorn kollar om en viss **position är tillgänglig** – dvs. inte en vägg och inom rutnätets gränser.

````python
# Sensor 3: Kan roboten röra sig till position (x, y)?
# -----------------------------------------------------
def can_move_to(x, y, world):
    """Returnerar True om positionen är inom gränser och inte en vägg."""
    # Kolla om positionen är utanför rutnätet
    if y < 0 or y >= len(world):        # Utanför raderna?
        return False
    if x < 0 or x >= len(world[0]):     # Utanför kolumnerna?
        return False
    # Kolla om cellen är en vägg
    if world[y][x] == WALL:
        return False
    return True  # Allt OK – roboten kan röra sig dit!

# Testa sensorn
print("=== Testa can_move_to ===")
print(f"Kan gå till (2,1) – tom cell?  {can_move_to(2, 1, world)}")  # Förväntat: True
print(f"Kan gå till (0,0) – hörnvägg?  {can_move_to(0, 0, world)}")  # Förväntat: False
print(f"Kan gå till (2,2) – innervägg? {can_move_to(2, 2, world)}")  # Förväntat: False
print(f"Kan gå till (4,1) – tom cell?  {can_move_to(4, 1, world)}")  # Förväntat: True
````

## Del 3: Kombinera sensorer – `and` och `or` 🔗

Ibland behöver roboten veta **flera saker samtidigt**.  

I Python kan vi kombinera `True`/`False` med:
- **`and`** – Båda måste vara `True`
- **`or`** – Minst en måste vara `True`
- **`not`** – Vänder på svaret (`True` → `False`)

```python
True and True   →  True
True and False  →  False
True or  False  →  True
not True        →  False
```

````python
# Kombinera sensorer
rx, ry = 3, 1  # Roboten på smutscell

# Ska roboten dammsuga?
# Ja, om det finns smuts OCH roboten kan dammsuga
finns_smuts = can_see_dirt(rx, ry, world)
print(f"Finns smuts på ({rx},{ry})?  {finns_smuts}")

# Är roboten hemma OCH finns det smuts?
hemma = is_at_home(rx, ry, home_x, home_y)
print(f"Är roboten hemma?            {hemma}")

kan_gå_höger = can_move_to(rx + 1, ry, world)
print(f"Kan roboten gå höger?        {kan_gå_höger}")

print()
print("--- Kombinationer ---")
print(f"Smuts OCH kan gå höger:      {finns_smuts and kan_gå_höger}")
print(f"Hemma ELLER smuts:           {hemma or finns_smuts}")
print(f"INTE hemma:                  {not hemma}")
````

## Del 4: Perception – allt samlat i ett objekt 📦

Istället för att anropa varje sensor separat kan vi samla all information i ett **dictionary** – en sorts lista med namngivna värden.

```python
{'dirt_here': True, 'at_home': False, ...}
```

Vi kallar det för **perception** – precis som hjärnans samlade bild av vad sinnen rapporterar.

````python
# Perception – samlar all sensordata i ett dictionary
# ----------------------------------------------------
def get_perception(robot_x, robot_y, world, home_x, home_y):
    """Returnerar en dictionary med allt roboten 'känner' just nu."""
    perception = {
        "dirt_here":    can_see_dirt(robot_x, robot_y, world),
        "at_home":      is_at_home(robot_x, robot_y, home_x, home_y),
        "can_move_right": can_move_to(robot_x + 1, robot_y, world),
        "can_move_left":  can_move_to(robot_x - 1, robot_y, world),
        "can_move_up":    can_move_to(robot_x, robot_y - 1, world),
        "can_move_down":  can_move_to(robot_x, robot_y + 1, world),
    }
    return perception

# Testa – vad 'känner' roboten på position (2, 3)?
rx, ry = 2, 3
p = get_perception(rx, ry, world, home_x, home_y)

print(f"Robotens perception på ({rx}, {ry}):")
for key, value in p.items():
    ikon = '✅' if value else '❌'
    print(f"  {ikon} {key}: {value}")
print()
visa_världen(world, rx, ry)
````
