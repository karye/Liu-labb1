## 🗺️ Del 2 – Vad är en "värld"?

En dator kan inte se bilder av ett hus. Men den kan läsa **siffror och tabeller**.

Vi beskriver ROBB:s värld som ett **rutnät** – precis som ett schackbräde!

Varje ruta i rutnätet kallas en **cell**. Varje cell innehåller information om vad som finns där.

### Exempel på ett litet rutnät (5×5):

```
# # # # #
# H . ░ #
# . # . #
# ░ . . #
# # # # #
```

Kan du läsa kartan? Här är förklaringen:

| Symbol | Emoji | Namn | Betydelse |
|--------|-------|------|-----------|
| `#`    | 🧱    | Vägg | Roboten kan INTE gå hit |
| `.`    | 🟫    | Tom  | Roboten KAN gå hit |
| `░`    | 🌫️   | Smuts | Roboten ska städa här |
| `H`    | 🏠    | Hem  | Robotens startposition |
| `R`    | 🤖    | Robot | Var roboten är just nu |

### Koordinater – var är cellen?

Vi använder **(x, y)** för att säga VAR i rutnätet vi är:
- **x** = kolumn (räknas från **vänster**, 0, 1, 2, ...)
- **y** = rad (räknas **uppifrån**, 0, 1, 2, ...)

```
       x=0  x=1  x=2  x=3  x=4
y=0  [  #    #    #    #    #  ]
y=1  [  #    H    .    ░    #  ]
y=2  [  #    .    #    .    #  ]
y=3  [  #    ░    .    .    #  ]
y=4  [  #    #    #    #    #  ]
```

Hemmet `H` är alltså på position **(x=1, y=1)**.
Den övre smutsen `░` är på position **(x=3, y=1)**.

## 🎨 Del 3 – Celltyper i koden

I Python representerar vi varje celltyp med ett **tal**. Det är lättare för datorn att jämföra siffror än bokstäver.

| Tal | Symbol | Emoji | Namn    |
|-----|--------|-------|---------|
| `0` | `.`    | 🟫    | Tom cell |
| `1` | `#`    | 🧱    | Vägg |
| `2` | `░`    | 🌫️   | Smuts |
| `3` | `H`    | 🏠    | Hem |
| `4` | `?`    | ❓    | Okänd (roboten har inte besökt cellen ännu) |

Det lilla rutnätet från del 2 ser ut så här i Python – som en **lista av listor**:

```python
# Rutnätet – en lista av rader, varje rad är en lista av celler
# Rad 0: [1, 1, 1, 1, 1]  → # # # # #
# Rad 1: [1, 3, 0, 2, 1]  → # H . ░ #
# Rad 2: [1, 0, 1, 0, 1]  → # . # . #
# Rad 3: [1, 2, 0, 0, 1]  → # ░ . . #
# Rad 4: [1, 1, 1, 1, 1]  → # # # # #

värld = [
    [1, 1, 1, 1, 1],
    [1, 3, 0, 2, 1],
    [1, 0, 1, 0, 1],
    [1, 2, 0, 0, 1],
    [1, 1, 1, 1, 1]
]
```

### Hur läser vi ut en cell?

Vi skriver `värld[y][x]` – **y-index kommer FÖRST**, sedan x-index.

```python
värld[1][1]  # → 3  (Hem)
värld[1][3]  # → 2  (Smuts)
värld[2][2]  # → 1  (Vägg)
```

## 💻 Del 4 – Vår första kod!

Nu är det dags att skriva riktig kod! 🎉

Vi börjar med **en enda funktion** som gör **en enda sak**:
den tar emot ett rutnät och skriver ut det på ett fint sätt.

### Vad är en funktion?

En funktion är som ett **recept**. Du:
1. Ger funktionen **ingredienser** (kallas *parametrar* eller *argument*)
2. Funktionen **gör något** med dem
3. Ibland **returnerar** den ett resultat

```
        ┌─────────────────────┐
 INPUT  │                     │  OUTPUT
──────► │    FUNKTION         │ ────────►
(värld) │  skriv_ut_värld()   │  (ingen, skriver bara ut)
        │                     │
        └─────────────────────┘
```

**Kör kodcellen nedan** genom att klicka på den och trycka **Shift + Enter** (eller klicka på ▶ knappen).

````python
# ============================================================
# FUNKTION 1: Omvandla ett cellnummer till ett tecken
# ============================================================
# Den här funktionen tar emot ett tal (0, 1, 2, 3 eller 4)
# och returnerar det tecken som motsvarar celltypen.

def cell_till_tecken(cell):
    # cell == 0 betyder: cellen är TOM
    if cell == 0:
        return "."   # Tom cell visas som en punkt

    # cell == 1 betyder: cellen är en VÄGG
    if cell == 1:
        return "#"   # Vägg visas som ett brädgård-tecken

    # cell == 2 betyder: cellen har SMUTS
    if cell == 2:
        return "░"   # Smuts visas som ett skuggat fyrkant

    # cell == 3 betyder: cellen är HEM
    if cell == 3:
        return "H"   # Hem visas som bokstaven H

    # cell == 4 betyder: cellen är OKÄND (roboten har inte sett den)
    if cell == 4:
        return "?"   # Okänd cell visas som ett frågetecken

    # Om vi inte känner igen talet – returnera ett felmeddelande
    return "!"


# ============================================================
# FUNKTION 2: Skriv ut hela världen på ett fint sätt
# ============================================================
# Den här funktionen tar emot ett rutnät (en lista av listor)
# och skriver ut det rad för rad.

def skriv_ut_värld(rutnät):
    # Vi lopar igenom varje rad i rutnätet
    # enumerate() ger oss både radnumret (y) och radens innehåll
    for y, rad in enumerate(rutnät):

        # Vi lopar igenom varje cell i raden
        for x, cell in enumerate(rad):

            # Omvandla cellen till ett tecken och skriv ut det
            # end=" " betyder att vi inte byter rad efter varje cell
            print(cell_till_tecken(cell), end=" ")

        # När vi är klara med en rad, byt till ny rad
        print()


# ============================================================
# Testa funktionerna med ett litet exempelrutnät
# ============================================================

# Här definierar vi ett litet rutnät (5×5)
# Varje siffra representerar en celltyp:
#  1 = Vägg, 3 = Hem, 0 = Tom, 2 = Smuts
exempel_värld = [
    [1, 1, 1, 1, 1],   # Rad 0: Övre väggen
    [1, 3, 0, 2, 1],   # Rad 1: Hem, tom, smuts
    [1, 0, 1, 0, 1],   # Rad 2: Tom, vägg, tom
    [1, 2, 0, 0, 1],   # Rad 3: Smuts, tom, tom
    [1, 1, 1, 1, 1],   # Rad 4: Nedre väggen
]

# Anropa funktionen – skriv ut världen!
print("Världen ser ut så här:")
print()
skriv_ut_värld(exempel_värld)
````

### 🔍 Vad hände?

Du borde se något som ser ut så här:

```
Världen ser ut så här:
```

Python läste igenom listan av listor och omvandlade varje siffra till ett tecken.

**Koden du just körde innehåller:**

### Experiment 5.2 – Skapa en större värld

Prova att ändra storleken på rutnätet! Nedan finns en funktion som **automatiskt** skapar ett rutnät av valfri storlek – med väggar runt om.

Du behöver **inte** förstå all kod i funktionen nu. Fokusera på vad du ändrar och vad som händer!

````python
# ============================================================
# FUNKTION: Skapa ett tomt rutnät
# ============================================================
# Den här funktionen tar emot en storlek (t.ex. 7 → 7×7 rutnät)
# och skapar automatiskt ett rutnät med väggar runt om och
# ett hem i övre vänstra hörnet innanför väggen.

def skapa_tomt_rutnät(storlek):
    # Skapa ett rutnät med bara tomma celler (0)
    # Det är en lista med 'storlek' stycken rader,
    # och varje rad har 'storlek' stycken tomma celler
    rutnät = [[0] * storlek for _ in range(storlek)]

    # Lägg till väggar (1) längs alla fyra kanter
    for i in range(storlek):
        rutnät[0][i] = 1          # Övre kanten
        rutnät[storlek-1][i] = 1  # Nedre kanten
        rutnät[i][0] = 1          # Vänstra kanten
        rutnät[i][storlek-1] = 1  # Högra kanten

    # Placera hem (3) i övre vänstra hörnet innanför väggen
    rutnät[1][1] = 3

    return rutnät  # Ge tillbaka det färdiga rutnätet


# ============================================================
# ✏️ UPPGIFT: Ändra storleken och se vad som händer!
# ============================================================
# Prova olika storlekar, t.ex.: 5, 7, 10, 15
# OBS: Minsta tillåtna storlek är 3 (annars ryms inte väggarna)

storlek = 7   # ← ÄNDRA DET HÄR TALET!

nytt_rutnät = skapa_tomt_rutnät(storlek)

print(f"Rutnät med storlek {storlek}×{storlek}:")
print()
skriv_ut_värld(nytt_rutnät)
print()
print(f"Antal rader: {len(nytt_rutnät)}")
print(f"Antal kolumner: {len(nytt_rutnät[0])}")
print(f"Totalt antal celler: {len(nytt_rutnät) * len(nytt_rutnät[0])}")
````

### Experiment 5.3 – Räkna celler

Nu ska du få skriva lite kod själv! Nedan finns en funktion som **räknar** hur många celler av en viss typ som finns i rutnätet.

Men – en del av koden saknas! Du behöver fylla i de tomma ställena (`___`).

````python
# ============================================================
# FUNKTION: Räkna celler av en viss typ
# ============================================================
# ✏️ UPPGIFT: Fyll i de tomma ställena (___) med rätt kod!
#
# Tips:
#  - Celltyper: 0=Tom, 1=Vägg, 2=Smuts, 3=Hem, 4=Okänd
#  - Använd 'räknare += 1' för att öka räknaren med 1
#  - Jämförelseoperatorn i Python är '==' (inte bara '=')

def räkna_celler(rutnät, celltyp):
    räknare = 0  # Vi börjar på 0

    # Gå igenom varje rad
    for rad in rutnät:
        # Gå igenom varje cell i raden
        for cell in rad:
            # Om cellen är av rätt typ, öka räknaren
            if cell == celltyp:
                räknare += 1  # ← Denna rad är klar (titta på den!)

    return räknare  # Ge tillbaka det slutliga antalet


# ============================================================
# Testa funktionen med exempelrutnätet
# ============================================================

test_värld = [
    [1, 1, 1, 1, 1],
    [1, 3, 0, 2, 1],
    [1, 0, 1, 0, 1],
    [1, 2, 0, 0, 1],
    [1, 1, 1, 1, 1],
]

# Skriv ut rutnätet så vi minns hur det ser ut
print("Rutnätet:")
print()
skriv_ut_värld(test_värld)
print()

# Räkna och skriv ut resultaten
antal_väggar = räkna_celler(test_värld, 1)   # Räkna väggar (celltyp 1)
antal_tomma  = räkna_celler(test_värld, 0)   # Räkna tomma celler (celltyp 0)
antal_smuts  = räkna_celler(test_värld, 2)   # Räkna smutsceller (celltyp 2)

print(f"🧱 Antal väggar:        {antal_väggar}")
print(f"🟫 Antal tomma celler:  {antal_tomma}")
print(f"🌫️  Antal smutsceller:   {antal_smuts}")
print()
print(f"Totalt antal celler: {len(test_värld) * len(test_värld[0])}")
````

### Experiment 5.4 – Visa roboten i världen

Nu ska vi visa ROBB:s position i rutnätet! Vi ger roboten en **(x, y)**-position och skriver ut `R` på den platsen.

````python
# ============================================================
# FUNKTION: Skriv ut världen MED roboten
# ============================================================
# Den här funktionen fungerar som skriv_ut_värld(), men tar
# också emot robotens position (robot_x, robot_y)
# och skriver ut 'R' på den platsen.

def skriv_ut_värld_med_robot(rutnät, robot_x, robot_y):
    for y, rad in enumerate(rutnät):
        for x, cell in enumerate(rad):

            # Är det här robotens position?
            if x == robot_x and y == robot_y:
                print("R", end=" ")  # Skriv ut R för roboten
            else:
                print(cell_till_tecken(cell), end=" ")  # Annars: vanlig cell

        print()  # Ny rad


# ============================================================
# ✏️ UPPGIFT: Ändra robotens position och se vad som händer!
# ============================================================
# OBS: Roboten kan inte stå på en vägg!
# Giltiga positioner (tomma eller hem-celler) är:
#   (1,1), (2,1), (3,1)
#   (1,2),        (3,2)
#   (1,3), (2,3), (3,3)

robot_x = 1   # ← ÄNDRA DET HÄR (0–4)
robot_y = 1   # ← ÄNDRA DET HÄR (0–4)

visa_värld = [
    [1, 1, 1, 1, 1],
    [1, 3, 0, 2, 1],
    [1, 0, 1, 0, 1],
    [1, 2, 0, 0, 1],
    [1, 1, 1, 1, 1],
]

print(f"Roboten är på position (x={robot_x}, y={robot_y}):")
print()
skriv_ut_värld_med_robot(visa_värld, robot_x, robot_y)
````
